# 2. 키워드 검색 추출

```python
import requests
import pandas as pd
from typing import Any

def search_key(query: str, dataframe: bool = False, page: int = 1, size: int = 15, **kwargs: Any) -> Any:
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": "KakaoAK " + "2d92e0d6158cce3f266723d7af43dd5a"}
    params = {"query": query, "page": page, "size": size}
    params.update(kwargs)  # kwargs를 params에 추가

    try:
        response = requests.get(url, headers=headers, params=params)
        response.raise_for_status()
    except Exception as e:
        print("Request Error")
        print(e)
        return

    try:
        if dataframe:
            return pd.DataFrame(response.json()["documents"])
        else:
            return response.json()
    except Exception as e:
        print("Export Error")
        print(e)
        return
```

```python
# 사용 예시
all_results = []

for page in range(1, 3):
    result = search_key("합천", dataframe=True, page=page)
    if result.empty:
        print(f"No results on page {page}")
        break
    all_results.append(result)
    print(f"Results for page {page}:\n", result)

# 모든 페이지의 결과를 하나의 DataFrame으로 합치기
final_results = pd.concat(all_results, ignore_index=True)
print(final_results)
```
